```python
import pandas as pd
%matplotlib inline
import numpy
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import seaborn

seaborn.set(style='white')

funds = numpy.sqrt([297000000, 37906178])
cp = seaborn.color_palette('Blues_r', n_colors=len(funds))

fig, ax = plt.subplots()

for n, p in enumerate(funds):

    circle = plt.Circle((1, p), radius=p, facecolor=cp[n])
    ax.add_artist(circle)

ax.set_xlim(-max(funds), max(funds))
ax.set_ylim(0, 2 * max(funds))
ax.set_aspect('equal')

plt.show()
```

```python
fig.savefig("fundinf_chart.pdf", transparent=False)
```
